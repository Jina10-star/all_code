```python
def int_split(a):
    a=str(a)
    if len(a)==3:
        primary_int=a 
        secondary_int=a
        tartiary_int=a
    else:
        d=a.split(',')
        if len(d)==2:
            primary_int=d[0]
            if len(d[1])==1:
                secondary_int=d[0]
                tartiary_int=d[1]
            else:
                secondary_int=d[1]
                tartiary_int=d[1]
        else:
            primary_int=d[0]
            secondary_int=d[1]
            tartiary_int=d[2]
        
    return primary_int,secondary_int,tartiary_int

    
```

```python
def cat_split(a):
    if ',' not in a:
        primary_cat=a 
        secondary_cat=a
        tartiary_cat=a
    else:
        g=a.split(',')
        #print(g)
        primary_cat=g[0]
        if len(g)==3:
            secondary_cat=g[1]
            tartiary_cat=g[2]
        else:
            if g[1] in ['TEMP_HEATED','TEMP_UNHEATED','TEMP_COLD']:
                secondary_cat=g[0]
                tartiary_cat=g[1]
            else:
                secondary_cat=g[1]
                tartiary_cat=g[1]
            
        
    return primary_cat,secondary_cat,tartiary_cat
```

```python
df2['primary_int']=df2['Integer'].map(lambda s:int_split(s)[0])
df2['secondary_int']=df2['Integer'].map(lambda s:int_split(s)[1])
df2['tartiary_int']=df2['Integer'].map(lambda s:int_split(s)[2])
df2['primary_cat']=df2['CAT_Name'].map(lambda s:cat_split(s)[0])
df2['secondary_cat']=df2['CAT_Name'].map(lambda s:cat_split(s)[1])
df2['tartiary_cat']=df2['CAT_Name'].map(lambda s:cat_split(s)[2])
df2['primary']=df2['primary_cat']+","+df2['primary_int']
df2['secondary']=df2['secondary_cat']+","+df2['secondary_int']
df2['tartiary']=df2['tartiary_cat']+","+df2['tartiary_int']
```
