```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
df = pd.read_csv("vvid_names.csv")
df = df[~df.parent_name.isin({'RiderItemDelivery', 'UberEATS MASTER', 'uberRUSH MASTER', 'eMobility MASTER'})]
df['name'] = df['name'].astype(str).apply(lambda r: r.replace(" ", ""))
text = " ".join(df['name'])
```
