```python
def levenshtein_distance(s1, s2):
    s1 = s1.lower()
    s2 = s2.lower()
    
    # Initialize a matrix to store the distances between substrings.
    dp = [[0] * (len(s2) + 1) for _ in range(len(s1) + 1)]
    
    # Initialize the first row and column.
    for i in range(len(s1) + 1):
        dp[i][0] = i
    for j in range(len(s2) + 1):
        dp[0][j] = j
    
    # Fill in the matrix.
    for i in range(1, len(s1) + 1):
        for j in range(1, len(s2) + 1):
            cost = 0 if s1[i - 1] == s2[j - 1] else 1
            dp[i][j] = min(dp[i - 1][j] + 1, dp[i][j - 1] + 1, dp[i - 1][j - 1] + cost)
    
    return dp[len(s1)][len(s2)]

def final_output_chatgpt(ground_truth,c):
    z=pd.DataFrame()
    for key in c.keys():
        if key in ground_truth:
            res1 = json.loads(ground_truth[key])
            x=c[key].replace("\n"," ")
            res2 =json.loads(x)
            total_entities = len(res1)
            correct_predictions = 0
            for entity, gt_value in res1.items():
                if entity in res2:
                    pred_value = res2[entity]
                    dist = levenshtein_distance(str(gt_value), str(pred_value))
                    if dist <= 5:
                        correct_predictions += 1
            accuracy = round((correct_predictions / total_entities) * 100,0)
            z=z.append({'file_name': key,'ground_truth':[res1],'chatgpt response':[res2],'accuracy':accuracy},ignore_index=True)
            z.to_csv(key+'.csv')
            return z
def extract_entities(row):
    p=row['chatgpt response'][0]
    #.replace("\n", " ")
   # chatgpt_response = json.loads(p)
    chatgpt_response = p
    return pd.Series(chatgpt_response)
```
