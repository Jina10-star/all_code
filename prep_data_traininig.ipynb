```python
from nltk.corpus import stopwords
sw = set(stopwords.words('english'))
```

```python
import re
```

```python
#lowering and removing punctuation
df['processed'] = df['email'].apply(lambda x: re.sub(r'[^\w\s]','', x.lower()))
#removing the numerical values and working only with text values
df['processed'] = df['processed'].apply(lambda x : re.sub('[^a-zA-Z]', " ", x ))
#removing the stopwords
df['processed'] = df['processed'].apply(lambda x: ' '.join([word for word in x.split() if word not in sw ]))
```

```python
import requests
```

```python
def get_intent(message):
    headers = {'Content-type': 'application/json'}
    payload = '{"text": "'+message+'","message_id": "a"}'
    raw_response = requests.post(
        "http://localhost:5005/model/parse",
        verify=False,data=payload,headers=headers).json()
    # raw_intent = raw_response
    # print('intent=',intent)
    # await asyncio.sleep(0.001)
#     print(raw_response)
    return raw_response['intent']['name'],raw_response['intent']['confidence']
```

```python
results = []
for index,row in test_set.iterrows():
#     print(row)
#     break
#     print(len(row))
    result = get_intent(row[5])
    results.append(result)
```

```python
test_set['predicted_intent'] = [result[0] for result in results]
test_set['predicted_confidence'] = [result[1] for result in results]
```

```python
wrong_predictions = 0
for index, row in test_set.iterrows():
    if test_set.loc[index,'Intent'].split() != test_set.loc[index,'predicted_intent'].split('_'):
        wrong_predictions+=1
#         print(os.path.basename(df.iloc[index,0]))
        print(test_set.loc[index,'Intent'], test_set.loc[index,'predicted_intent'], test_set.loc[index,'predicted_confidence'])
```
