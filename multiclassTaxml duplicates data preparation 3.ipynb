```python
import pandas as pd
```

```python
df_integers = pd.read_excel('../data/TrainingDataSet_Integer.xlsx')
```

```python
df_duplicates = df_integers[df_integers.duplicated(subset=['item_name','description','establishment_type'],keep=False)]
```

```python
sorted_duplic_df = df_duplicates.sort_values(by=['item_name','description','establishment_type'], ascending=True)
```
