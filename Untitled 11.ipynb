```python
def check_level(Credit_Score):
    if Credit_Score<300:
        p="low"
    elif Credit_Score>500:
        p="high"
    else:
        p="moderate"
    return p
```

```python
df['category_new']=df['Credit_Score'].map(check_level)
```
