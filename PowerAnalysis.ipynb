```python
pa = SwitchbackPowerAnalysis(
    switch_granularity=SWITCHING,
    segment_filters={
        'city': CITY_LIST_LATAM,
        'product': PRODUCT_LIST,
    },
    #segment_types=['city'],
    metric_names=X_METRICS,
    start_time=START_TIME
)
```

```python
pa.collect_data()
pa.run_analysis() 
df = pa.mde_table()
```
